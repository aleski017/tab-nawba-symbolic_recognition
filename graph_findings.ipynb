We employ the library called **Graphmuse** to load scores fromm XML to Pytorch graphs

```python
import partitura
import torch
from utilities.constants import *
import graphmuse as gm

track = '0fb722f6-35a3-436b-9316-4c2de2b86cc7'
score = partitura.load_musicxml(DF_PATH_TRACKS + track + '/' + track + '.xml')
note_array = score.note_array()
```

```python
feature_array = torch.rand((len(note_array), 10)) 
import torch
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# Convert note_array into feature matrix
feature_list = []

for note in note_array:
    pitch = note[2]       # Example: MIDI pitch
    duration = note[1]    # Example: duration in beats
    start_time = note[0]  # Example: onset time
    velocity = note[3]    # Example: loudness
    measure = note[4]     # Example: measure number
    beat_pos = note[5]    # Example: position in measure
    time_sig = note[6]    # Example: time signature ID
    key_sig = note[7]     # Example: key signature ID
    instrument = 0        # If single instrument, otherwise assign real value
    chord_label = 0       # If chord information is available, otherwise 0
    
    # Append extracted features
    feature_list.append([pitch, duration, start_time, velocity, measure, beat_pos, time_sig, key_sig, instrument, chord_label])

feature_list = np.array(feature_list)

scaler = MinMaxScaler()
feature_list = scaler.fit_transform(feature_list)
# Convert to tensor
feature_array = torch.tensor(feature_list, dtype=torch.float32)

score_graph = gm.create_score_graph(feature_array, note_array)
simple_graph = score_graph.to_homogeneous()
#print(simple_graph)
simple_graph.to_dict()
```

```python
x_dict_score = {'note' : simple_graph.to_dict()['x']}
edge_index_dict_score = {('note', 'onset', 'note'): simple_graph.to_dict()['edge_index']}
```

```python
import graphmuse.nn as gmnn
import torch

# Define the number of input features, output features, and edge features
num_input_features = 10
num_hidden_features = 10
num_output_features = len(MIZAN_LIST_TRAIN)
num_layers = 1
# metadata needs to be provided for the metrical graph similarly to Pytorch Geometric heterogeneous graph modules.
metadata = (
    ['note'],
    [('note', 'onset', 'note')]
)

# Create an instance of the MetricalGNN class
metrical_gnn = gmnn.MetricalGNN(num_input_features, num_hidden_features, num_output_features, num_layers, metadata=metadata)

# Create some dummy data for the forward pass
num_nodes = 5
x_dict = {'note': torch.rand((num_nodes, num_input_features))}
edge_index_dict = {('note', 'onset', 'note'): torch.tensor([[0, 1, 2, 3, 4], [1, 2, 3, 4, 0]])}

# Perform a forward pass
out = metrical_gnn(x_dict_score, edge_index_dict_score)
print(edge_index_dict)
print(x_dict)
```
